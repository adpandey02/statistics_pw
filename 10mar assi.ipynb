```python
Q1: Estimation statistics is a branch of statistics that deals with estimating the characteristics of a
population based on a sample. Point estimate is a single value that is used to estimate the population parameter, 
while interval estimate is a range of values that is used to estimate the population parameter.
```

```python
# ans 2
import math

def estimate_population_mean(sample_mean, sample_std, sample_size):
    z_score = 1.96  # for 95% confidence interval
    margin_of_error = z_score * (sample_std / math.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return (lower_bound, upper_bound)
```

```python
Q3: Hypothesis testing is a statistical method used to determine whether a hypothesis about a population parameter is true or not. 
It is used to make inferences about the population based on the sample data. The importance of hypothesis testing is 
that it allows us to draw conclusions about the population based on a sample, and to test whether the results are statistically significant.


Q4: H0: The average weight of male college students is greater than the average weight of female college students.
```

```python
from scipy import stats

# sample data
sample1 = [10, 15, 12, 14, 18]
sample2 = [8, 11, 9, 13, 10]

# conduct two-sample t-test
t_stat, p_val = stats.ttest_ind(sample1, sample2)

# print results
print("t-statistic:", t_stat)
print("p-value:", p_val)
```
