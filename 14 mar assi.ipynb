```python
Q1
import scipy.stats as stats

def calculate_f_value(data1, data2):
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)
    
    f_value = var1 / var2
    df1 = len(data1) - 1
    df2 = len(data2) - 1
    
    p_value = stats.f.cdf(f_value, df1, df2)
    
    return f_value, p_value
```

```python
Q2
import scipy.stats as stats

def calculate_critical_f_value(significance_level, df1, df2):
    alpha = significance_level / 2
    critical_value = stats.f.ppf(1 - alpha, df1, df2)
    
    return critical_value
```

```python
Q3
import numpy as np
import scipy.stats as stats

def perform_f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    
    f_value = var1 / var2
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
    
    p_value = stats.f.cdf(f_value, df1, df2)
    
    return f_value, df1, df2, p_value

# Generate random samples from two normal distributions
sample1 = np.random.normal(loc=0, scale=1, size=100)
sample2 = np.random.normal(loc=0, scale=1.5, size=100)

# Perform F-test
f_value, df1, df2, p_value = perform_f_test(sample1, sample2)

print("F-value:", f_value)
print("Degrees of freedom:", df1, ",", df2)
print("p-value:", p_value)
```

```python
Q4
import numpy as np
import scipy.stats as stats

def perform_f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    
    f_value = var1 / var2
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
    
    p_value = stats.f.cdf(f_value, df1, df2)
    
    return f_value, df1, df2, p_value

# Generate random samples from two normal distributions with known variances
sample1 = np.random.normal(loc=0, scale=np.sqrt(10), size=12)
sample2 = np.random.normal(loc=0, scale=np.sqrt(15), size=12)

# Perform F-test
f_value, df1, df2, p_value = perform_f_test(sample1, sample2)

alpha = 0.05
critical_value = stats.f.ppf(1 - alpha / 2, df1, df2)

print("F-value:", f_value)
print("Degrees of freedom:", df1, ",", df2)
print("p-value:", p_value)
print("Critical F-value:", critical_value)

if f_value > critical_value:
    print("Reject the null hypothesis. Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis. Variances are not significantly different.")
```

```python
Q5
import numpy as np
import scipy.stats as stats

def perform_f_test(sample, claimed_variance):
    var = np.var(sample, ddof=1)
    f_value = var / claimed_variance
    df1 = len(sample) - 1
    df2 = len(sample)
    
    p_value = stats.f.cdf(f_value, df1, df2)
    
    return f_value, df1, df2, p_value

# Sample data and claimed variance
sample = np.random.normal(loc=0, scale=np.sqrt(0.006), size=25)
claimed_variance = 0.005

# Perform F-test
f_value, df1, df2, p_value = perform_f_test(sample, claimed_variance)

alpha = 0.01
critical_value = stats.f.ppf(1 - alpha / 2, df1, df2)

print("F-value:", f_value)
print("Degrees of freedom:", df1, ",", df2)
print("p-value:", p_value)
print("Critical F-value:", critical_value)

if f_value > critical_value:
    print("Reject the null hypothesis. The claim about the variance is not justified.")
else:
    print("Fail to reject the null hypothesis. The claim about the variance is justified.")
```
